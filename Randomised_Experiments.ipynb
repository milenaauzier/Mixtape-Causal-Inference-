##**Experimentos Randomizados**

Expeimentos randomizados atribuem aleatoriamente indivíduos em uma população a um tratamento ou a um grupo de controle. A proporção que recebe o tratamento não precisa ser de 50%. Você pode realizar um experimentos em apenas 10% de suas amostras que receberam o tratamento. 

fonte: Causal Inferece for the Brave and True

**A randomização aniquila o viés ao tornar os resultados potenciais independentes do tratamento.** 

Vamos entender o que significa "resultados potenciais independente do tratamento":

"Em resumo, a ideia dos resultados potenciais independentes do tratamento é que, para inferir causalidade, precisamos comparar os resultados observados com os resultados que teriam sido obtidos em diferentes condições de tratamento. Isso ajuda os pesquisadores a entender se um tratamento tem um efeito real e a identificar os fatores que afetam o sucesso de uma intervenção." ChatGTP

"O resultado potencial é como o resultado teria sido sob tratamento (Y1) ou sob controle (Y0). Em estudos randomizados, não queremos que o resultado seja independente do tratamento, pois pensamos que o tratamento causa o resultado. Mas queremos que os resultados potenciais sejam independentes do Tratamento."

Abaixo um exemplo em python, no qual o experimento acontenceu baseado na seguinte situação: 

1. Alunos que receberam aulas online
2. Alunos que receberam aulas pelo metodo tradicional em sala de aula 
3. Alunos que receberam aulas online e em sala de aula

Os Alunos foram aleátoriamente escolhidos para que não ocorrece viéis, e assim obervarmos apenas a causalidade do experimento. 

*   format_ol = 0 e 1 (alunos que receberam aulas on-line)
*   format_blended = 0.0 e 1.0 (alunos que receberam aulas em formato misturado)
*   face_to_face = 0 e 1 (para alunos que receberam aulas em sala de aula)
                           

```python
import pandas as pd
import numpy as np

data = pd.read_csv("/content/online_classroom.csv")
print(data.shape)
data.head()
```

```python
(data
 .assign(class_format = np.select(
     [data["format_ol"].astype(bool), data["format_blended"].astype(bool)],
     ["online", "blended"],
     default="face_to_face"
 ))
 .groupby(["class_format"])
 .mean())
```

```python
result = 78.54 - 73.63
result 
```

Podemos obeservar que as aulas presenciais têm média de 78,54 equanto os cursos online têm média de 73,63. Isso significa que o aprendizado online é negativo em -4,91. 
Você não precisa se preocupar que os cursos online possam ter alunos mais pobres que não podem pagar aulas presenciais ou, nesse caso, você não precisa se preocupar que os alunos dos diferentes tratamentos sejam diferentes de alguma forma além do tratamento que receberam. Por design, o experimento aleatório é feito para eliminar essas diferenças.

Por esse motivo, uma boa verificação de sanidade para ver se a randomização foi feita corretamente (ou se você está olhando para os dados corretos) é verificar se os tratados são iguais aos não tratados nas variáveis ​​de pré-tratamento.

**Experimentos randomizados ou ensaios controlados randomizados (RCT) são a maneira mais confiável de obter efeitos causais. É uma técnica direta e absurdamente convincente.**
